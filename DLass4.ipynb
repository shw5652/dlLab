```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-5, 5, 0.01)
```

```python
def plot(func, yaxis=(-1.4, 1.4)):
    plt.ylim(yaxis)
    plt.locator_params(nbins=5)
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    plt.axhline(lw=1, c='black')
    plt.axvline(lw=1, c='black')
    plt.grid(alpha=0.4, ls='-.')
    plt.box(on=None)
    plt.plot(x, func(x), lw=3)
```

1) Sigmoid Activation Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plot(sigmoid, yaxis=(-0.4, 1.4))
```

2) Tanh Activation Function

```python
def tanh(x):
    return 2 / (1 + np.exp(-2 * x)) -1
```

```python
plot(tanh)
```

3) ReLU Activation Function

```python
relu = np.vectorize(lambda x: x if x > 0 else 0, otypes=[float])
```

```python
plot(relu, yaxis=(-0.4, 1.4))
```

4) Leaky ReLU Activation Function

```python
leaky_relu = np.vectorize(lambda x: max(0.1 * x, x), otypes=[float])
```

```python
plot(leaky_relu)
```

5) ELU Activation Function

```python
elu = np.vectorize(lambda x: x if x > 0 else 0.5 * (np.exp(x) - 1), otypes=[float])
```

```python
plot(elu)
```

6) Softmax Activation Function

```python
# def softmax(x):
#   e_x = np.exp(x - np.max(x))
#   return e_x / np.sum(e_x, axis=0)
```

```python
# plot(softmax, yaxis=(-0.4, 1.4))
```

```python
def softmax(x):
  """It takes a vector of real numbers as input and outputs a probability distribution over multiple classes."""
  e_x = np.exp(x - np.max(x))
  return e_x / np.sum(e_x, axis=0)
```

```python
def plot_softmax():
    class_labels = ["Class A", "Class B", "Class C"]
    probabilities = np.array([0.2, 0.3, 0.5])
    plt.bar(class_labels, probabilities)
    plt.xlabel("Classes")
    plt.ylabel("Probability")
    plt.title("Softmax Activation Function")
    plt.show()
```

```python
z = np.array([1, 2, 3])
result = softmax(z)
print(result)
```

```python
plot_softmax()
```

7) Softplus Activation Function

```python
def softplus(x):
    return np.log(1+np.exp(x))
```

```python
plot(softplus, yaxis=(-0.4, 1.4))
```

8) Binary Step Activation Function

```python
binary_step = np.vectorize(lambda x: 1 if x > 0 else 0, otypes=[float])
```

```python
plot(binary_step, yaxis=(-0.4, 1.4))
```

9) Piecewise Linear Activation Function

```python
piecewise_linear = np.vectorize(lambda x: 1 if x > 3 else 0 if x < -3 else 1/6*x+1/2, otypes=[float])
```

```python
plot(piecewise_linear, yaxis=(-0.4, 1.4))
```
